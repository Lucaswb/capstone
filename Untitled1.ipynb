```python
from PIL import Image
import imutils
import os
import timeit
import shutil
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gray_scale(query, end_query, width = 200):
    """this will imput a file directory and go through that file directory and select all the png files within the directory, 
    resizes the images and converts them to a grey scale and then saves them to a file """
    #creates a list of all files ending with png"
    file_list = list(filter(lambda x: x != '.DS_Store', os.listdir(query)))
    
    
    #clears out the altered_cards directory
    for root, dirs, files in os.walk(end_query):
        for f in files:
            os.unlink(os.path.join(root, f))
        for d in dirs:
            shutil.rmtree(os.path.join(root, d))

    
    for name in file_list:
        img_query = query + name # creates the string to query the image and converts to greyscale
        print(img_query)
        img = cv2.imread(img_query, 0)
        print(img.shape[0], img.shape[1])
        
        #resize image
        w = width
        h = int((img.shape[0]/img.shape[1])*w)
        resized = cv2.resize(img, (w,h), interpolation = cv2.INTER_AREA)
        save_query = end_query + name
        cv2.imwrite(save_query, resized)
    
    return
        
    
    
```

```python
def warp(img, values = 100):
    cd_img = np.asarray(img.copy())
    h = cd_img.shape[0]
    w = cd_img.shape[1] 
    mask = np.asarray(Image.new('L', (w,h), 255))

    c_width, c_height = cd_img.shape

    persX1 = np.random.randint(1, values)
    persX2 = np.random.randint(1, values)
    persX3 = np.random.randint(1, values)
    persX4 = np.random.randint(1, values)
    persY1 = np.random.randint(1, values)
    persY2 = np.random.randint(1, values)
    persY3 = np.random.randint(1, values)
    persY4 = np.random.randint(1, values)


    pts1 = np.float32([[0,0],[c_width,0],[0,c_height],[c_width,c_height]])
    pts2 = np.float32([[0,persY2],[c_width-persX2,persY2],[persX3,c_height-persY3],[c_width-persX4,c_height-persY4]])

    M = cv2.getPerspectiveTransform(pts1,pts2)
    dst = cv2.warpPerspective(cd_img,M,(c_height,c_width))
    dst_mask = cv2.warpPerspective(mask,M,(c_height,c_width))
    
    
    return dst, dst_mask
    
```

```python
def rotate(img):
    ranint = np.random.randint(0,360)
    rotated = imutils.rotate_bound(img, ranint)
    return rotated
```

```python
def cr_train_img(img_query, bg_query, card_warp = 20,bg_width = 1000):

    #loads image as a greyscale
    image = cv2.imread(img_query, 0)
    mask = Image.new('L', image.shape, 255)
    w_card, w_mask = warp(image, values = card_warp)
    
#     #this will be commented out once images load properly
#     resized_warped = cv2.resize(warped, (150,200), interpolation = cv2.INTER_AREA)

    # creates a PIL image after warping
    img = Image.fromarray(w_card)
    mask = Image.fromarray(w_mask)
    
    #loads random background

    bg_list = list(filter(lambda x: x != '.DS_Store', os.listdir(bg_query)))
    bg_rand = np.random.choice(bg_list)
    
    bg_file = bg_query + bg_rand
    bg = Image.open(bg_file, 'r')
    
    # resize background images while maintaining aspect ratio
    bg_w, bg_h = bg.size[0], bg.size[1] 
    bg_height = int(bg_h/bg_w *bg_width) 
    bg = bg.resize((bg_width , bg_height), Image.ANTIALIAS)

    

    random_height = np.random.randint(0, bg.size[1]-img.size[1])
    random_width = np.random.randint(0, bg.size[0]-img.size[0])
    random_rotate = np.random.randint(0, 360)


#     mask = Image.new('L', img.size, 255)
    img = img.rotate(random_rotate, expand = True)
    mask = mask.rotate(random_rotate, expand =True)
    

    text_img = Image.new('L', bg.size, 0)
    text_img.paste(bg, (0,0))
    text_img.paste(img, (random_width,random_height), mask = mask)
    text_img
    return text_img
```

```python
imageQ = 'images/altered_cards/2ofDiamonds.png'
bgQ = 'images/backgroundnew/'

cr_train_img(imageQ, bgQ)
```

```python
start = timeit.timeit()
cd_query = 'images/altered_cards/'
bgQ = 'images/backgroundnew/'
save_path = 'images/saved_images/'

card_list = [item for item in os.listdir(cd_query) if ".png" in item]

for card in card_list:
    imageQ = cd_query + card
    img = cr_train_img(imageQ, bgQ)
    save_query = save_path + card
    img.save(save_query, format = 'png')
end = timeit.timeit()
print(end - start)
     
```
